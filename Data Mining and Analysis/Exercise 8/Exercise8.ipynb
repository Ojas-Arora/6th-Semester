```python
import pandas as pd
import numpy as np
data=pd.read_csv('./Colleges.csv')
data.head(10)
```

Drop ones with no Qualifying Names

```python
nullNames=data[data['School'].isnull()].index.tolist()
data.drop(nullNames, inplace=True)
data.head(102)
```

Clean the Numeric Attributes

```python
def clean(item):
  if type(item) is float and np.isnan(item):
    return np.nan
  if type(item) is float:
    return item
  if item.find('*')!=-1:
    if item.find('/')!=-1:
      SAT=item[:item.find('*')]
      ACT=item[item.find('/')+1:item.find('*', item.find('*')+1)]
      return max(int(SAT), int(ACT))
    else:
      return int(item[:item.find('*')])
  if item.find('/')!=-1:
    SAT=item[:item.find('/')]
    ACT=item[item.find('/')+1:item.find('%')]
    return max(int(SAT), int(ACT))
  if(item.find('%')):
    return int(item[:item.find('%')])
data['*SAT or ACT']=data['*SAT or ACT'].apply(clean)
data['*SAT or ACT'].fillna(data['*SAT or ACT'].median(), inplace=True)
data.head(20)
```

Clean cost based Attributes

```python
costBased=[]
for attribute in list(data):
  if type(data[attribute][0]) is str and data[attribute][0][0]=='$':
    costBased.append(attribute)
def removeDollarOrComma(item):
  if type(item) is float and np.isnan(item):
    return np.nan
  item.strip()
  item=item[1:]
  item=float(''.join(item.split(',')))
  return item
for attribute in costBased:
  data[attribute]=data[attribute].apply(removeDollarOrComma)
  data[attribute].fillna(data[attribute].median(), inplace=True)
data.head(102)
```

Clean other Rates

```python
rateBased=[]
for attribute in list(data):
  if type(data[attribute][0]) is str and data[attribute][0][-1]=='%':
    rateBased.append(attribute)
def removePercent(item):
  if (type(item) is str and item[0].isalpha()) or (type(item) is float and np.isnan(item)):
    return np.nan
  item.strip()
  item=item[:-1]
  return float(item)
for attribute in rateBased:
  data[attribute]=data[attribute].apply(removePercent)
  data[attribute].fillna(data[attribute].median(), inplace=True)
data.head(102)
```
